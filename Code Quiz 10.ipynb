```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

```python
hw10 = pd.read_csv('homework_10.1.csv', index_col=0)
hw10.head()
```

```python
# Fixed effect for each month
model = smf.ols('y ~ C(time)', data=hw10).fit()
print(model.summary())
model.params.filter(like='C(time)')
```

```python
# Fixed effect for each city
model = smf.ols('y ~ C(city)', data=hw10).fit()
print(model.summary())
model.params.filter(like='C(city)')
```

```python
num = 10000000

X = np.clip(np.random.normal(3, 1, (num,)), 0.01, 100)
Z = np.clip(np.random.normal(3, 1, (num,)), 0.01, 100)
Y = np.log(X + Z) + np.random.normal(0, 1, (num,))
df = pd.DataFrame({'X': X, 'Z': Z, 'Y': Y})
df.head()
```

```python
model = smf.ols('np.exp(Y) ~ X + Z', data=df).fit()
model.summary()
```

```python
num=10000
Z = np.random.normal(0, 1, (num,))
X = Z + np.random.normal(0, 1, (num,))
Y = 1.5 * X + 2.3 * Z + np.random.normal(0, X**2, (num,))
df = pd.DataFrame({'X': X, 'Z': Z, 'Y':Y})
```

```python
model = smf.ols('Y ~ X + Z', data=df).fit()
model.summary()
```

```python
coef = []
for n in range(100):
    num=10000
    Z = np.random.normal(0, 1, (num,))
    X = Z + np.random.normal(0, 1, (num,))
    Y = 1.5 * X + 2.3 * Z + np.random.normal(0, X**2, (num,))
    df = pd.DataFrame({'X': X, 'Z': Z, 'Y':Y})

    model = smf.ols('Y ~ X + Z', data=df).fit()
    coef.append(model.params.iloc[1])

np.std(coef)
```
